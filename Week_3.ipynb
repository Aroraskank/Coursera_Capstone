# Part-3
## Folium Map to show the neighbourhood of Toronto

```python
import folium

map_toronto = folium.Map(location=[(df["Latitude"][0]), (df["Longitude"][0])], zoom_start=10)

for lat, lng, community, neighborhood in zip(df['Latitude'], df['Longitude'], df['Community'], df['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, community)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    

map_toronto
```
